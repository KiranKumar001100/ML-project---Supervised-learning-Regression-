# SVM

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
x = df[['car_ID', 'symboling', 'doornumber', 'wheelbase','carlength', 'carwidth', 'carheight', 'curbweight', 'cylindernumber',
        'enginesize', 'boreratio', 'stroke', 'compressionratio', 'horsepower', 'peakrpm', 'citympg', 'highwaympg']]
# Target variable
y = df['price']
```

```python
# Split the data set into training and testing
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
# Train a Gradient Boosting Regressor
gb_regressor = GradientBoostingRegressor()
gb_regressor.fit(x_train, y_train)
```

```python
# Make predictions with the Gradient Boosting model
gb_predictions = gb_regressor.predict(x_test)
```

```python
x_combined = np.column_stack((x_test, gb_predictions))
```

```python
# Create the Support Vector Machine Regressor (SVM)
svm_regressor = SVR(kernel = 'linear', C = 1.0)
```

```python
# Fit the SVM model
svm_regressor.fit(x_combined, y_test)
```

```python
# Predict the values of the test data using the SVM
svm_pred = svm_regressor.predict(x_combined)
```

```python
svm_pred
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
mse3 = mean_squared_error(y_test, svm_pred)
print('Mean Squared Error (MSE):', mse3)

rmse3 = np.sqrt(mse3)
print('Root Mean Squared Error (RMSE):',rmse3 )

mae3 = mean_absolute_error(y_test, svm_pred)
print('Mean Absolute Error (MAE):', mae3)

r_square3 = r2_score(y_test, svm_pred)
print('R-squared (coefficient of determination):',r_square3)
```

```python
# Scatter plot
plt.figure(figsize=(20, 10))
plt.scatter(y_test, svm_pred, color = 'blue', label = 'Predicted')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color = 'red', label = 'Perfect Prediction')
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Actual vs Predicted Prices')
plt.legend()
plt.show()
```
